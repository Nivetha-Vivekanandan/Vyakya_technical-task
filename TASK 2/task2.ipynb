```python
import datetime
def date_validation(day, month, year):
     
    isValidDate = True
     
    try :
        datetime.datetime((year),
                          (month), (day))
         
    except ValueError :
        isValidDate = False
         
    if(isValidDate) :
        print ("Yes")
    else :
        print ("No")
 
date_validation(data_out['DAY'],data_out['MONTH'],data_out['YEAR'])
```
